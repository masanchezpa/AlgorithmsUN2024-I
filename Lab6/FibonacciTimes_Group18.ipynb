# 1) Plot of the recursive Fibonacci execution time for n in [0,50]

```
# fibR - Fibonacci Recursive O( ( (1+sqrt(5)/2 )^n lg n ) [Relaxed asymptotic bound O(2^n lg n)]
def fibR(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)    # for large n´s  the addition will take lg n stetp because the numbers are very large
```

```
#Plot of the recursive Fibonacci execution time for n in [0,45]
import time
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(46)
y = []
for n in x:
    start_time = time.time()
    fibR(n)
    y.append(time.time()-start_time)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Recursive Fibonacci')
plt.plot(x, x, label='y=n', linestyle='--')

# Adding labels, title, and legend
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Recursive Fibonacci Execution Time')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

# 2) Plot of  the iterative Fibonacci linear and iterative Fibonacci logarithmic execution time n in [0,1000]

```
# fibI - Fibonacci Iterative (forward) O(n lgn)
def fibI(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```
# Define the range of n values
x = np.arange(1001)

# Measure the execution time of fibI for each n
y = []
for n in x:
    start_time = time.time()
    fibI(n)
    y.append(time.time() - start_time)

# Plot the execution times
plt.plot(x, y, label='Iterative Fibonacci')

# Add y=n to the plot for comparison
plt.plot(x, x, label='y=n')

# Set labels and title
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Execution Time of Iterative Fibonacci Function')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

# 3) Plot of Fibonacci memoization (dynamic programming) algorithm (backward)execution time n in [0,1000]

```
# fibI - Fibonacci Memoization (dynamic programming) (backward) O(n lg n)

def fibM(n):
    memo = {}
    def fib(n):
        if n==0 :
            return 0
        if n==1 :
            return 1
        if (n-2) not in memo:
            memo[n-2] = fib(n-2)
        if (n-1) not in memo:
            memo[n-1] = fib(n-1)
        return memo[n-1] + memo[n-2]
    return fib(n)
```

```
# Define the range of n values
x = np.arange(1001)

# Measure the execution time of fibM for each n
y = []
for n in x:
    start_time = time.time()
    fibM(n)
    y.append(time.time() - start_time)

# Create a larger figure
plt.figure(figsize=(10, 6))

# Plot the execution times with a thicker line
plt.plot(x, y, label='Logarithmic Fibonacci', linestyle='-', linewidth=2)

# Add y=n to the plot for comparison with a different line style
plt.plot(x, x, label='y=n', linestyle='--', linewidth=2)

# Set labels and title
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Execution Time of Logarithmic Fibonacci Function')


# Add grid lines
plt.grid(True, which='both')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
